```python
def generar_individuos(numero_individuos, numero_genes):
  # Crear una lista vacía para almacenar todas las listas
  individuos = []

  # Crear 20 listas con 3 números aleatorios cada una
  for i in range(numero_individuos):
      cromosomas = []
      for j in range(numero_genes):
          nuevo_gen = random.randint(0, 9)
          cromosomas.append(nuevo_gen)
      individuos.append(cromosomas)

  return individuos

def mostrar_individuos(poblacion):
  # Imprimir todas las listas generadas
  print(f"numero de individuos: {len(poblacion)}")
  for individuo in poblacion:
    print(individuo)
```

```python
def main():
  poblacion_actual=int(input("Cantidad de poblacion actual"))
  mutaciones= int (input("Escojer cada cuantas generaciones se generar mutaciones"))
  ciclo_de_mutacion=1 
  mejores=int(poblacion_actual/2)
  poblacion = generar_individuos(poblacion_actual,mejores)
  generacion=0
  while True:
    #mostrar_individuos(poblacion)
    generacion+=1
    print("======GENRACION",generacion,"============")
    mejores_individuos = obtener_mejores_individuos(poblacion , mejores)
    mostrar_individuos(mejores_individuos)
    if mejores_individuos[0][10]==0:
       print("solucion : ", mejores_individuos[0])
       break
    mejores_individuos_sin_fitness = [fila[:10] for fila in mejores_individuos]
    nueva_poblacion = generar_nuevos_individuos(mejores_individuos_sin_fitness)#crear nuevos 100 individuos
    if ciclo_de_mutacion == mutaciones:
       print("---------------------Se genero mutaciones ---------------")
       poblacion = mutacion(nueva_poblacion)
       ciclo_de_mutacion = 0
    else:
       ciclo_de_mutacion += 1
       poblacion = nueva_poblacion

      
```

## funcion de mutacion : Seleccionar un idividuo segun una probabilidad y modifica alazar en una posicion , con valore aleatorios entre 0-10

```python
def mutacion(poblacion):
  probabilidad_de_mutacion=0.15
  for i in range (len(poblacion)):
    if random.random()< probabilidad_de_mutacion:#seleccionar un idividuo alazar para mutar
       poblacion[i][random.randint(0,9)]= random.randint(0,10)
  return poblacion 
```
